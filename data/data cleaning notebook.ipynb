```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
final_student_data = pd.read_csv("final_student_data.csv")
```

```python
final_student_data
```

```python
final_student_data.info()
```

```python
total_students = final_student_data.shape[0]

# Total number of male students
total_male_students = final_student_data[final_student_data["Gender"] == "Male"].shape[0]

# Total number of female students
total_female_students = final_student_data[final_student_data["Gender"] == "Female"].shape[0]

# Print the results
print("Total number of students: ", total_students)
print("Total number of male students: ", total_male_students)
print("Total number of female students: ", total_female_students)
```

```python
final_student_data["Year Of Study"] = final_student_data["Year Of Study"].str.title()
```

```python
final_student_data
```

```python
final_student_data["Hostel"] = final_student_data["Hostel"].str.title()
```

```python
final_student_data["Hostel"].unique()
```

```python
final_student_data["Hostel"].fillna("Day Scholar", inplace = True)
```

```python
final_student_data["Hostel"].replace({'No':'Day Scholar', 'Yes': 'Hostel'}, inplace=True)
```

```python
final_student_data["Hostel"].unique()
```

```python
final_student_data.to_csv("final_student_data.csv", index=False)
```

```python
discontinued = final_student_data[final_student_data["Discontinued Flag"] == 1].shape[0]

# Get the number of students who were retained
retained = final_student_data[final_student_data["Retained Flag"] == 1].shape[0]

# Get the number of students who moved between departments
moved = final_student_data[final_student_data["Moved Between Department Flag"] == 1].shape[0]

print("Number of students who discontinued:", discontinued)
print("Number of students who were retained:", retained)
print("Number of students who moved between departments:", moved)
```
